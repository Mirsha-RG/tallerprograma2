```python
# primero crear una función que nos ayude a ver si
# dentro de un grupo específico de n personas 2 o más tienen el mismo cumpleaños

import random # importamos random para generar números aleatorios
import matplotlib.pyplot as plt

# checar que en el mismo grupo alguien cumple años
def cumple_grupo(grupo):
  cumpleanios = set() # se usa set para crear un conjunto y que solo guarde un único cumpleaños
  for _ in range(grupo): # para cada persona en el grupo
    cumple = random.randint(1,365) # con random genero los cumpleaños de mis "personas" del día 1 al 365 en un año no bisiesto
    if cumple in cumpleanios: # si el cumple está en mi conjunto de cumpleaños entonces
      return True
    cumpleanios.add(cumple)
  return False

# simular - repetir un mismo evento n veces
def simulacion(tam_grupo, repeticiones):
  conteo = 0
  for _ in range(repeticiones):
    if cumple_grupo(tam_grupo):
      conteo +=1
  return conteo / repeticiones

# testear grupos de multiplos de 5

eje_x = []
eje_y = []
for grupo in range(5,101,5):
  prob = simulacion(grupo,1000)
  eje_x.append(grupo)
  eje_y.append(prob)
  print(f"La probabilidad de que en un grupo de {grupo} personas haya al menos 2 personas que comparten cumpleaños es {prob:.2%}")

plt.plot(eje_x,eje_y,color='red', linewidth=3)
plt.title("Paradoja del Cumpleaños")
plt.xlabel("Tamaño del grupo")
plt.ylabel("Probabilidad")









```
